```python
#Answer_3

Flask class has a redirect() function. When called, it returns a response object and 
redirects the user to another target location with specified status code. location parameter
is the URL where response should be redirected.

from flask import *  
   
app = Flask(_name_) 
@app.route('/admin')  
def admin():  
    return 'admin'  
  
@app.route('/student')  
def student():  
    return 'student'  
  
@app.route('/user/<name>')  
def user(name):  
    if name == 'admin':  
        return redirect(url_for('admin'))  
    if name == 'student':                            #redirect() method is send to user to different url 
        return redirect(url_for('student'))  
if _name_ =='_main_':  
    app.run(debug = True)
```
